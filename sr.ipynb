```python
import math
import matplotlib.pyplot as plt

# === AI-GENERATED FUNCTION ===
def nearest_neighbor(cities):
    tour = [cities[0]]
    unvisited = cities[1:]
    while unvisited:
        last = tour[-1]
        next_city = min(unvisited, key=lambda c: math.dist(last, c))
        tour.append(next_city)
        unvisited.remove(next_city)
    return tour

# === MANUAL IMPROVEMENTS ===
def validate_cities(cities):
    if not cities or len(cities) < 2:
        raise ValueError("At least 2 cities required for a tour.")
    if not all(isinstance(c, tuple) and len(c) == 2 for c in cities):
        raise TypeError("Cities must be tuples of (x, y).")

# === PLOTTING FUNCTION ===
def plot_tour(tour):
    x, y = zip(*tour)
    plt.figure(figsize=(8,6))
    plt.plot(x, y, 'o-', color='blue', label='Tour Path')
    plt.plot([x[-1], x[0]], [y[-1], y[0]], 'r--', label='Return to Start')  # optional return to start
    for i, city in enumerate(tour):
        plt.text(city[0]+0.1, city[1]+0.1, str(i), fontsize=12)
    plt.title("Nearest Neighbor TSP Tour")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.grid(True)
    plt.show()

# === DEMO ===
def run_demo():
    cities = [(0,0), (2,3), (5,2), (6,6), (8,3)]
    validate_cities(cities)           # manual validation
    tour = nearest_neighbor(cities)   # AI-generated logic

    print("Tour order:", tour)
    plot_tour(tour)                   # plot the tour

if __name__ == "__main__":
    run_demo()
```
